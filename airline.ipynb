# Get the sum of null values for each column

```python
missValues = df.isnull().sum()
# missValues
for col, miss in missValues.items():
    print(f"Column '{col}': {miss} missing values")
```

## print and Drop the dublicated values.

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

# fill the null values with median for numeric values

```python
# df.fillna(df.median()[0], inplace=True) 
# df.median()
# df.median(numeric_only=True)
df.fillna(df.median(numeric_only=True), inplace=True)
```

# fill the null values with the mode for object values

```python
df.fillna(df.select_dtypes(include=['object']).mode().iloc[0], inplace=True)
# iloc[0] → Takes the first mode value (in case of multiple modes).
```

# new dataframe 

```python
airline_df = df[['Year', 'Flights', 'Reporting_Airline']]
airline_df.head()
```

# Get the summation for each year and Reporting_Airline

```python
unique_airline = airline_df.groupby(['Year', 'Reporting_Airline'])['Flights'].sum().reset_index()
unique_airline.head()
# Using .reset_index() ensures that the index is reset to a default numerical order (0,1,2...).
```

# Initialize Dash app

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
app = dash.Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1("Airline Flights Dashboard", style={'text-align': 'center','color': 'white'}),

    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(year), 'value': year} for year in unique_airline['Year'].unique()],
        placeholder="Select a Year",
        value=unique_airline['Year'].min()
    ),

    dcc.Graph(id='flights-bar-chart')
])

# Callback to update graph
@app.callback(
    Output('flights-bar-chart', 'figure'),
    Input('year-dropdown', 'value')
)
def update_chart(selected_year):
    filtered_df = unique_airline[unique_airline['Year'] == selected_year]

    fig = px.bar(
        filtered_df,
        x='Reporting_Airline',
        y='Flights',
        title=f'Total Flights per Airline in {selected_year}',
        labels={'Flights': 'Number of Flights', 'Reporting_Airline': 'Airline'},
        color='Reporting_Airline'
    )

    return fig

if __name__ == '__main__':
    app.run_server(port=8003, host='127.0.0.1',debug=True)
```

```python
airline_data = df[["Year", "Flights", "Reporting_Airline", "OriginState"]]

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Airline Dashboard", style={'text-align': 'center', "font-size": 40,'color':'white'}),

    # Input for Year
    html.Div([
        html.Span("Enter Year: ", style={'color': 'white', "font-size": 25}),
        dcc.Input(id='input-yr', value=2010, type='number', style={'height': '50px', "font-size": 25})
    ], style={"font-size": 25}),

    html.Br(),

    # Input for Origin State
    html.Div([
        html.Span("Origin State: ", style={'color': 'white', "font-size": 25}),
        dcc.Input(id='input-state', type='text', placeholder="e.g., CA", style={'height': '50px', "font-size": 25})
    ], style={"font-size": 25}),

    html.Br(),

    # Graph output
    html.Div(dcc.Graph(id='bar-plot'))
])

# Define callback function
@app.callback(
    Output(component_id='bar-plot', component_property='figure'),
    [Input(component_id='input-yr', component_property='value'),
     Input(component_id='input-state', component_property='value')]
)
def get_graph(entered_year, entered_state):
    entered_year = int(entered_year)  # Ensure year is an integer

    # Filter dataset based on year
    df_filtered = airline_data[airline_data['Year'] == entered_year]

    # Apply Origin State filter if user entered a state
    if entered_state:
        df_filtered = df_filtered[df_filtered['OriginState'].str.lower() == entered_state.lower()]

    if df_filtered.empty:
        return px.bar(title=f"No data available for {entered_year} in {entered_state}")

    # Aggregate data
    g1 = df_filtered.groupby('Reporting_Airline')['Flights'].sum().nlargest(10).reset_index()

    # Create bar chart
    fig1 = px.bar(g1, x='Reporting_Airline', y='Flights',
                  title=f"Top 10 Airlines in {entered_year} ({entered_state}) by Number of Flights")

    fig1.update_layout(xaxis_title="Airline", yaxis_title="Number of Flights")

    return fig1

# Run the app
if __name__ == '__main__':
    app.run_server(port=8004, host='127.0.0.1', debug=True)
```

```python
airline_data2 = df[["Year", "Month", "Flights", "CarrierDelay", "WeatherDelay", "Reporting_Airline", "OriginState"]]

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div(style={'background-color': 'black', 'color': 'white', 'padding': '10px'}, children=[
    html.H1("Launch Application", style={'text-align': 'center', "font-size": 40, 'color': 'white'}),

    html.Br(),

    # the body
    html.Div([
        html.Label("Input Year:", style={'color': 'white', "font-size": 20, 'font-weight': 'bold'}),
        dcc.Dropdown(
            id='year-dropdown', #the input_ID
            options=[{'label': str(year), 'value': year} for year in range(2005, 2021)], # 2021 exclude
            placeholder="Select a Year",
            value=2005,  # default year
            style={"background-color": "white","color": "black",}
        )
    ], style={"margin-bottom": "20px"}),

    # Graphs side by side
    html.Div([
        # Carrier Delay Graph
        html.Div(dcc.Graph(id='carrier-delay-plot'), style={'width': '50%', 'display': 'inline-block'}),
        
        # Weather Delay Graph
        html.Div(dcc.Graph(id='weather-delay-plot'), style={'width': '50%', 'display': 'inline-block'})
    ])
])

# callback function
@app.callback(
    [Output('carrier-delay-plot', 'figure'),
    Output('weather-delay-plot', 'figure')],
    [Input('year-dropdown', 'value')]
)
def update_charts(selected_year):
    # Filter dataset based on selected year
    df_filtered = airline_data[airline_data['Year'] == selected_year]

    # Handle empty data
    if df_filtered.empty:
        return px.line(title=f"No data available for {selected_year}"), px.line(title=f"No data available for {selected_year}")

    # Monthly Average Carrier Delay
    carrier_delay = df_filtered.groupby(['Month', 'Reporting_Airline'])['CarrierDelay'].mean().reset_index()
    fig1 = px.line(
        carrier_delay,
        x='Month',
        y='CarrierDelay',
        color='Reporting_Airline',
        title="Average Carrier Delay Time (minutes) by Airline"
    )

    # Monthly Average Weather Delay
    weather_delay = df_filtered.groupby(['Month', 'Reporting_Airline'])['WeatherDelay'].mean().reset_index()
    fig2 = px.line(
        weather_delay,
        x='Month',
        y='WeatherDelay',
        color='Reporting_Airline',
        title="Average Weather Delay Time (minutes) by Airline"
    )

    return fig1, fig2


# Run the app
if __name__ == '__main__':
    app.run_server(port=8005, host='127.0.0.1', debug=True)
```
